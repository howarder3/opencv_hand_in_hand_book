### 1-1-1 檢查 matplotlib等相關套件是否有正常安裝

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import glob
```

### 1-1-2 撰寫一個 matplotlib的涵式，使得我們之後的結果能直接在 jupyter中直接顯示

```python
def show_img(img):
    image_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(image_rgb)
    plt.show()
```

## 1-3 利用OpenCV從檔案讀取圖片

```python
file_name = "./cat.jpg"  # 指定你要的圖片路徑
origin_img = cv2.imread(file_name) # 讀取圖片 load image
show_img(origin_img) # 使用我們在 1-1-2 小節定義好的 show_img 函式
```

## 1-4 利用OpenCV顯示圖片、開啟與關閉顯示圖片的視窗

```python
file_name = "./cat.jpg"  # 指定你要的圖片路徑
origin_img = cv2.imread(file_name) # 讀取圖片 load image
cv2.imshow('Image window', origin_img) # 顯示圖片，第一個參數表示視窗名稱，第二個參數就是你的圖片

cv2.waitKey(0) # 暫停等待按鍵，使 cv2.imshow 能顯示出畫面
cv2.destroyAllWindows() # 配合上一行，按下任意鍵則關閉所有視窗
```

## 1-5 利用OpenCV將圖片存檔

```python
file_name = "./cat.jpg"  # 指定你要讀取的圖片路徑
origin_img = cv2.imread(file_name) # 讀取圖片 load image
cv2.imwrite('./out.jpg', origin_img) # 儲存圖片，第一個參數為圖片路徑(可直接修改副檔名)，第二個為圖片
```

## 1-5-1延伸用法，使用OpenCV進行圖片轉檔

```python
file_name = "./cat.jpg"  # 指定你要讀取的圖片路徑
origin_img = cv2.imread(file_name) # 讀取圖片 load image
cv2.imwrite("./cat.png", origin_img) # 儲存圖片，第一個參數為圖片路徑(可直接修改副檔名)，第二個為圖片
```
